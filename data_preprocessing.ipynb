```python
import os 
import pandas as pd
import numpy as np
```

```python
# Get the current directory
current_directory = os.getcwd()

print(f"current directory: {current_directory}")
```

```python
file = f"{current_directory}/datasets/fredkellner/onthesnow-ski-area-reviews/versions/1/OnTheSnow_SkiAreaReviews.csv"
```

```python
df = pd.read_csv(file, header=0)
df
```

```python
# Set the first column ('col1') as the index
df = df.drop(df.columns[0], axis=1)  # Drop by index
df
```

```python
# impute the missing value in Reviewer Name with anonymous_user
df['Reviewer Name'].fillna("anonymous_user", inplace=True)
df
```

```python
df.columns
```

```python
# add a new columns to indicate missing reviews
def generate_review_text_boolean(row):
    if row['Review Text'] is np.nan:
        return 0
    else:
        return 1
    
    
# if row['Review Text'] is np.nan else 0
df['Review Missing']=df.apply(generate_review_text_boolean,axis=1)
df
```
